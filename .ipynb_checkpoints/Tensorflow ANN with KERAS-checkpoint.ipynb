```python
class_names = ["t-shirt/top", "Trouser", "Pullover", "Dress", "Coat", "Sandal", "Shirt", "Sneaker", "Bag", "Ankle Boot"]
```

```python
#PREPROCESSING
train_images = train_images/255  #255 is the pixel max value, so everything ill convert from 0 to 1(Doing this so that processing time for correcting values reduces reduces)
test_images = test_images/255
```

```python
#MODEL BUILD
model = keras.Sequential([
    keras.layers.Flatten(input_shape = (28, 28)), #Input layer 1 FLATTEN because we will enter 2d array(28 x 28) and it will cinvert that to 1d array(748)
    keras.layers.Dense(128, activation = 'relu'), #Hidden layer 2 with 128 neurnons.
    keras.layers.Dense(10, activation = 'softmax') #Output layer
])
```

```python
#COMPILE THE MODEL
model.compile(optimizer = 'adam',       #Version og gradient decent.
             loss = 'sparse_categorical_crossentropy',   # loss function lise cos function to calC entropy/LOSS.
             metrics = ['accuracy'])  #Outputs of metrics
```

```python
#TRAINING
model.fit(train_images, train_labels, epochs = 10) #Accuracy  > 90% , but if training accuracy is higher than eval accuracy, it's OVERFITTING OF DATA.
```

```python
#EVALUATION OF MODEL
test_loss, test_accuracy = model.evaluate(test_images, test_labels,verbose = 1) #verbose is 'are we looking at output or not'
```

```python
#PREDICTION
pred = model.predict(test_images)
```
